```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.columns
```

```python
# 1.	Barcha yo’lovchilar soni, jumladan erkaklar soni, ayollar soni aniqlansin.
all_passengers = df.shape[0]
male_passengers = df[df.Sex == 'male'].shape[0]
female_passengers = df[df.Sex == 'female'].shape[0]
print(all_passengers)
print(male_passengers)
print(female_passengers)
```

```python
# 2.	Erkaklar va ayollar soniga mos pie chart hosil qilinsin.
plt.pie([male_passengers, female_passengers], labels=["male_passengers","female_passengers"])
```

```python
# 3.	Titanic kemasiga chiqish biletlari narxining umumiy summasi va bitta bilet narxining o’rtacha qiymati aniqsin.
summa = sum(df.Fare)
average = summa/df.Fare.shape[0]
print(summa)
print(average)
```

```python
# 4.	15-40 yosh oralig’idagi yo’lovchilar umumiy yo’lovchilarning necha foizini tashkil etishi aniqlansin.
above = df[df.Age >= 15]
below = above[above.Age<=40]
percent = (below.shape[0]/df.shape[0])*100
print(percent)
```

```python
# 5.	Tirik qolgan bolalar (yoshi 16 dan kichik) barcha bolalarning necha foizini tashkil etishi aniqlansin. 
children = df[df.Age<=16].shape[0]
survived_children = df[(df.Age<=16) & (df.Survived == 1)].shape[0]
percent_surv_children = (survived_children/children)*100
print(percent_surv_children)
```
